### grafico de una circunferencia con munpy

```python
import matplotlib.pyplot as plt
import numpy as np

# Crear el gráfico
fig, ax = plt.subplots()

# Establecer los límites del plano cartesiano
ax.set_xlim(-10, 10)
ax.set_ylim(-10, 10)

# Parámetros del lente
radio_curvatura = 1.5
espesor = 0.5

# Función para calcular el perfil del lente convergente
def perfil_lente_convergente(x):
    return np.sqrt(radio_curvatura**2 - x**2)#+ espesor /2

# Función para calcular el perfil del lente convergente invertido
def perfil_lente_convergente_invertido(x):
    return -np.sqrt(radio_curvatura**2 - x**2)#- espesor/2

# Rango de valores en el eje X
x = np.linspace(-radio_curvatura, radio_curvatura, 1000)

# Perfil del lente convergente
perfil_convergente = perfil_lente_convergente(x)

# Perfil del lente convergente invertido
perfil_convergente_invertido = perfil_lente_convergente_invertido(x)

# Dibujar el lente convergente
ax.plot(x, perfil_convergente, color='blue', label='Lente Convergente')
# Dibujar el lente convergente invertido
ax.plot(x, perfil_convergente_invertido, color='blue', label='Lente Convergente Invertido')

# Mostrar leyenda y etiquetas de los ejes
ax.legend()
ax.set_xlabel('Posición en el eje X')
ax.set_ylabel('Posición en el eje Y')

# Mostrar el gráfico
plt.show()
```

#### wedge
La clase Wedge se utiliza para representar una cuña en un gráfico circular. Se 
define por su centro, radio, ángulo de inicio y ángulo final.

wedge = mpatches.Wedge(center, radius, theta1, theta2, color='magenta')

- center: Especifica las coordenadas (x, y) del centro de la cuña.
- radius: Define el radio de la cuña.
- theta1: Indica el ángulo de inicio en grados.
- theta2: Indica el ángulo final en grados.

set_aspect('equal') garantiza que el gráfico se muestre como un círculo.

```python
import matplotlib.pyplot as plt
import matplotlib.patches as mpatches

fig, ax = plt.subplots()

# Crear un Wedge
center = (0.5, 0.5)  # Coordenadas del centro de la cuña (x, y)
radius = 0.3  # Radio de la cuña
theta1 = 45  # Ángulo de inicio en grados
theta2 = 135  # Ángulo final en grados

wedge = mpatches.Wedge(center, radius, theta1, theta2, color='magenta')

# Agregar el Wedge al gráfico
ax.add_patch(wedge)

# Configurar límites y aspecto
ax.set_xlim(0, 1)
ax.set_ylim(0, 1)
ax.set_aspect('equal')  # Para asegurar que se grafique como un círculo

plt.title('Ejemplo de Wedge')
plt.grid(True)
plt.show()
```

### subplot 
 un subplot se refiere a la división de una figura o gráfico en varias subfiguras más pequeñas. La biblioteca matplotlib es comúnmente utilizada para crear subplots en Python.

```python
import matplotlib.pyplot as plt

# Datos de ejemplo
x = [1, 2, 3, 4]
y1 = [1, 4, 9, 16]
y2 = [1, 8, 27, 64]

# Crear los subplots
fig, (ax1, ax2) = plt.subplots(2)

# Subplot 1
ax1.plot(x, y1)
ax1.set_title('Subplot 1')
ax1.set_xlabel('X')
ax1.set_ylabel('Y1')

# Subplot 2
ax2.plot(x, y2)
ax2.set_title('Subplot 2')
ax2.set_xlabel('X')
ax2.set_ylabel('Y2')

# Mostrar los subplots
plt.tight_layout()
plt.show()

'''
 En este ejemplo, se crea una figura que contiene dos subfiguras (subplots). Cada
 subfigura se representa mediante un objeto Axes en matplotlib, que se almacena en
 las variables ax1 y ax2. Luego, se trazan las líneas correspondientes a los datos 
 y1 e y2 en cada subfigura, y se establecen los títulos y etiquetas adecuados para
 cada eje.

 El método plt.subplots(2) crea una figura con 2 subplots apilados verticalmente.
 Si se desea una disposición diferente, se pueden especificar otros argumentos en
 subplots(), como plt.subplots(1, 2) para dos subplots en una fila o
 plt.subplots(2, 2) para una matriz de 2x2 subplots, entre otras opciones.
 '''
```

##### como graficar vectores

```python
import matplotlib.pyplot as plt

def grafica_vectores():
    # Coordenadas de los vectores
    x = [1]
    y = [2]

    # Crear una figura y ejes
    fig, ax = plt.subplots()

    # Graficar los vectores
    ax.quiver(0, 0, x, y, angles='xy', scale_units='xy', scale=1, color = 'green')

    # Configurar límites y aspecto del gráfico
    ax.set_xlim([-1, 3])
    ax.set_ylim([-1, 3])
    ax.set_aspect('equal')

    #eje "x" y eje "y"
    plt.axvline(x=0, color='k', zorder=10)  # Línea vertical
    plt.axhline(y=0, color='k', zorder=10)  # Línea horizontal

    # Etiquetas y título
    ax.set_xlabel('Eje x')
    ax.set_ylabel('Eje y')
    ax.set_title('Vectores en el plano')

    # Mostrar el gráfico
    plt.grid(True)
    plt.show()

# Llamar a la función para graficar los vectores
grafica_vectores()

```

```python
import matplotlib.pyplot as plt

def rayos_incidentes(f, d, h, foco,imagen):
    if h != 0:
        #rayo paralelo
        plt.plot([0, d], [h, h], color = 'red', linewidth = 1)

        # pendiente del rayo que pasa por el foco
        x1, y1 = [0, foco[0]], [h, 0]
        m = (y1[1] - y1[0]) / (x1[1] - x1[0])
        # rayo que pasa por el foco
        x2 = [0, d]
        y2 = [h, h + m * (x2[1] - x2[0])]
        plt.plot(x2, y2, color='green', linewidth=1)

        #recta que pasa por el centro
        plt.plot([0, d], [h, 0], color='blue', linewidth=1)

    else:
        if d == f:
            plt.plot([0, d], [0, 0], color='red', linewidth=1)
            plt.plot([0, d], [0, 14], color='green', linewidth=1)
            plt.plot([0, d], [0, -14], color='blue', linewidth=1)

    #eje "x" y eje "y"
    plt.axvline(x=0, color='k', zorder=10)  # Línea vertical
    plt.axhline(y=0, color='k', zorder=10)  # Línea horizontal
    
    #punto focal
    plt.plot(d-f, 0, marker = 'o', color = 'k')
    plt.plot(d+f, 0, marker = 'o', color = 'k')

    #agrego rayos emergentes
    rayos_emergentes(f, d, h, foco,imagen)
    plt.show()

def rayos_emergentes(f, d, h, foco, imagen):
    magnificacion = d/imagen
    altura_imagen = h*magnificacion
    if h != 0:
        #rayo que pasa por el foco emerge paralelo
        plt.plot([d, d+imagen], [-h, -h], color = 'green', linewidth = 1)

        # pendiente del rayo que emerge por el foco
        x1, y1 = [0, foco[0]], [h, 0]
        m = -(y1[1] - y1[0]) / (x1[1] - x1[0])
        # rayo que emerge por el foco
        x2 = [d, d + imagen]
        y2 = [h, h - m * (x2[1] - x2[0])]
        plt.plot(x2, y2, color='red', linewidth=1)

        #recta que pasa por el centro sigue por el centro hata la imagen
        plt.plot([d, d+imagen], [0, -h], color='blue', linewidth=1)

    else:
        if d == f:
            plt.plot([d, d+f], [14, 0], color='red', linewidth=1)
            plt.plot([d, d+f], [0, 0], color='green', linewidth=1)
            plt.plot([d, d+f], [-14, 0], color='blue', linewidth=1)

    #punto focal
    plt.plot(d-f, 0, marker = 'o', color = 'k')
    plt.plot(d+f, 0, marker = 'o', color = 'k')
    plt.show()

f = 5#int(input('ingrese el foco de el lente\n')) #distancia focal
d = 10#int(input('a que distancia esta el lente del objeto?\n')) #a que distancia esta el lente?
h = 3#int(input('a que altura respecto al eje axial esta ubicado el objeto?\n')) #altura de el lente
foco = [d-f, 0] #coordenada del foco
imagen = 1/((1/f)-(1/d))

rayos_incidentes(f, d, h, foco, imagen)
```
