```python
import matplotlib.pyplot as plt
import matplotlib.patches as mpatches

def lente_divergente():
    fig, ax = plt.subplots()

    # Crear Wedges
    wedge_1 = mpatches.Wedge((-57, 0), 55, 0, 360, color='white', alpha = 1)
    wedge_2 = mpatches.Wedge((57, 0), 55, 0, 360, color='white')

    #crear primer cuadrado
    cuadrado_1 = plt.Polygon([[-20,-24.8], [-20,24.8], [20,24.8], [20,-24.8]], color = 'blue')

    # Agregar el Wedge al gráfico
    ax.add_patch(cuadrado_1)
    ax.add_patch(wedge_1)
    ax.add_patch(wedge_2)

    # # Recortar una cuña con la otra para mostrar la intersección
    # wedge_2.set_clip_path(cuadrado_1)
    # wedge_1.set_clip_path(cuadrado_1)

    # Configurar límites y aspecto
    ax.set_xlim(-100, 100)
    ax.set_ylim(-100, 100)
    ax.set_aspect('equal')  # Para asegurar que se grafique como un círculo

    plt.title('uso de lente divergente')
    plt.grid(True)
    plt.show()


def lente_convergente():
    fig, ax = plt.subplots()

    # Crear Wedges
    wedge_1 = mpatches.Wedge((-49, 0), 55, 0, 360, color='magenta', alpha = 0)
    wedge_2 = mpatches.Wedge((49, 0), 55, 0, 360, color='blue')
    
    # Agregar el Wedge al gráfico
    ax.add_patch(wedge_1)
    ax.add_patch(wedge_2)

    # Recortar una cuña con la otra para mostrar la intersección
    wedge_2.set_clip_path(wedge_1)

    # Configurar límites y aspecto
    ax.set_xlim(-100, 100)
    ax.set_ylim(-100, 100)
    ax.set_aspect('equal')  # Para asegurar que se grafique como un círculo

    plt.title('uso de lente convergente')
    plt.grid(True)
    plt.show()

lente_convergente()
lente_divergente()
```

### modificacion de las funciones


```python
import matplotlib.pyplot as plt
import matplotlib.patches as mpatches

def grafica_lente_divergente():
    fig, ax = plt.subplots()

    try:
        # Cambiar de posicion los lentes
        d = int(input("cuanto se desplazo el lente?\n").strip())
    except ValueError:
        print('esta muy equivocado señor')
        exit()

    # Crear Wedges
    wedge_1 = mpatches.Wedge((-57+d, 0), 55, 0, 360, color='white', alpha = 1)
    wedge_2 = mpatches.Wedge((57+d, 0), 55, 0, 360, color='white')

    #crear cuadrado
    cuadrado_1 = plt.Polygon([[-20+d,-24.8], [-20+d,24.8], [20+d,24.8], [20+d,-24.8]], color = 'blue')

    # Agregar el Wedge al gráfico
    ax.add_patch(cuadrado_1)
    ax.add_patch(wedge_1)
    ax.add_patch(wedge_2)

    # Agregar las líneas con alto valor de zorder
    plt.axvline(x=0, color='k', zorder=10)  # Línea vertical
    plt.axhline(y=0, color='k', zorder=10)  # Línea horizontal

    # Configurar límites y aspecto
    ax.set_xlim(-10, 300)
    ax.set_ylim(-50, 50)
    ax.set_aspect('equal')  # Para asegurar que se grafique simetricamente

    plt.title('sistema de lente divergente.')
    plt.grid(True)
    plt.show()


def grafica_lente_convergente():
    fig, ax = plt.subplots()

    try:
        # Cambiar de posicion los lentes
        d_1 = int(input("cuanto se desplazo el lente?\n").strip())
    except ValueError:
        print('esta muy equivocado señor')
        exit()

    # Crear Wedges{cuñás}
    wedge_1 = mpatches.Wedge((-49+d_1, 0), 55, 0, 360, color='magenta', alpha = 0)
    wedge_2 = mpatches.Wedge((49+d_1, 0), 55, 0, 360, color='aquamarine')
    
    # Agregar el Wedge al gráfico
    ax.add_patch(wedge_1)
    ax.add_patch(wedge_2)

    # Agregar las líneas con alto valor de zorder
    plt.axvline(x=0, color='k', zorder=10)  # Línea vertical
    plt.axhline(y=0, color='k', zorder=10)  # Línea horizontal

    # Recortar una cuña con la otra para mostrar la intersección
    wedge_2.set_clip_path(wedge_1)

    # Configurar límites y aspecto
    ax.set_xlim(-10, 300)
    ax.set_ylim(-50, 50)
    ax.set_aspect('equal')  # Para asegurar que se grafique simetricamente

    plt.title('sistema de lente convergente.')
    plt.grid(True)
    plt.show()

def grafica_par_de_lentes():    
    fig, ax = plt.subplots()

    try:
        # distancia del primer lente
        d_c = int(input("a que distancia esta el primer lente del objeto?\n").strip())

        # distancia del segundo lente
        d_d = int(input("a que distancia esta el segundo lente del objeto?\n ").strip())

    except ValueError:
        print('esta muy equivocado señor')
        exit()

    # componentes lente convergente
    wedge_1 = mpatches.Wedge((-49+d_c, 0), 55, 0, 360, color='magenta', alpha = 0)
    wedge_2 = mpatches.Wedge((49+d_c, 0), 55, 0, 360, color='aquamarine')

    # Agregar el Wedge al gráfico
    ax.add_patch(wedge_1)
    ax.add_patch(wedge_2)

    # Recortar una cuña con la otra para mostrar la intersección
    wedge_2.set_clip_path(wedge_1)

    # componentes de el lente divergente
    wedge_3 = mpatches.Wedge((-57+d_d, 0), 55, 0, 360, color='white', alpha = 1)
    wedge_4 = mpatches.Wedge((57+d_d, 0), 55, 0, 360, color='white', alpha = 1)

    #cuadrado componente de el lente divergente
    cuadrado_1 = plt.Polygon([[-8+d_d,-24.8], [-8+d_d,24.8], [8+d_d,24.8], [8+d_d,-24.8]], color = 'blue', alpha = 0.8)

    # Agregar el Wedge al gráfico
    ax.add_patch(cuadrado_1)
    ax.add_patch(wedge_3)
    ax.add_patch(wedge_4)
    ax.add_patch(wedge_1)
    ax.add_patch(wedge_2)

    #eje "x" y eje "y"
    plt.axvline(x=0, color='k', zorder=10)  # Línea vertical
    plt.axhline(y=0, color='k', zorder=10)  # Línea horizontal


    # Configurar límites y aspecto
    ax.set_xlim(-10, 300)
    ax.set_ylim(-50, 50)
    ax.set_aspect('equal')  # Para asegurar que se grafique simetricamente

    plt.title('sistema compuesto de dos lentes')
    plt.grid(True)
    plt.show()

try: 
    lentes = int(input('con que lentes desea trabajar, con uno [1] o con dos [2]:\n').strip())

    if lentes == 1:
        con_div =  int(input('va a trabajar con lentes convergentes [1] o divergentes[2]:\n').strip())
        if con_div == 1:
            grafica_lente_convergente()

        elif con_div == 2:
            grafica_lente_divergente()

        else:
            print('respuesta inadecuada: ')
            exit()

        print('esta muy equivocado señor')

    elif lentes == 2:
        grafica_par_de_lentes()

except ValueError:
    print('esta muy equivocado señor')
    exit()


```
